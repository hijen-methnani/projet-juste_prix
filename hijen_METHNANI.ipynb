```python
import numpy as np
import random

def bot_juste_prix_casino(a=1, b=1000):
    nombre_secret = random.randint(a, b)
    essais_max = 7  
    mises = [0, 10, 20, 40, 80, 160, 320]  
    gains = [1000, 800, 600, 400, 200, 100, 50]  
    gain_perte = 10  
    total_mise = 0
    
    borne_min, borne_max = a, b
    
    for essai in range(essais_max):
        proposition = (borne_min + borne_max) // 2  
        total_mise += mises[essai]
        
        if proposition == nombre_secret:
            gain_brut = gains[essai]
            gain_net = gain_brut - total_mise
            print("Résultat : JOUEUR GAGNE")
            print(f"Nombre secret : {nombre_secret}")
            print(f"Trouvé à l'essai : {essai+1}")
            print(f"Gain brut joueur : {gain_brut} €")
            print(f"Mises engagées : {total_mise} €")
            print(f"Gain net joueur : {gain_net} €")
            print(f"Profit casino : {-gain_net} €")  
            return gain_net   
        elif proposition < nombre_secret:
            borne_min = proposition + 1
        else:
            borne_max = proposition - 1
    
    gain_brut = gain_perte
    gain_net = gain_brut - total_mise
    print("Résultat : CASINO GAGNE")
    print(f"Nombre secret : {nombre_secret}")
    print(f"Mises engagées par le joueur : {total_mise} €")
    print(f"Gain de consolation : {gain_brut} €")
    print(f"Perte nette joueur : {-gain_net} €")
    print(f"Profit casino : {-gain_net} €")
    return gain_net

def calculer_esperance_casino(n=1000):
    print(f"Simulation de {n} parties pour le casino...")
    gains_nets_joueur = []
    for i in range(n):
        gain_joueur = bot_juste_prix_casino(1, 1000)
        gains_nets_joueur.append(gain_joueur)
    esperance_joueur = np.mean(gains_nets_joueur)
    profits_casino = [-gain for gain in gains_nets_joueur]
    esperance_casino = np.mean(profits_casino)
    profit_total_casino = np.sum(profits_casino)
    
    print("=" * 60)
    print("RAPPORT FINANCIER DU CASINO:")
    print(f"Espérance de gain du JOUEUR : {esperance_joueur:.2f} € (négative = bon pour le casino)")
    print(f"Espérance de profit du CASINO : {esperance_casino:.2f} € par partie")
    print(f"Profit total du casino sur {n} parties : {profit_total_casino:.2f} €")
    print(f"Nombre de parties gagnées par le joueur : {sum(1 for g in gains_nets_joueur if g > 0)}")
    print(f"Nombre de parties gagnées par le casino : {sum(1 for g in gains_nets_joueur if g <= 0)}")
    print(f"Taux de victoire du casino : {sum(1 for g in gains_nets_joueur if g <= 0) / n * 100:.1f}%")
    
    return esperance_casino, profit_total_casino

print("TEST D'UNE PARTIE:")
bot_juste_prix_casino(1, 1000)

print("\nCALCUL DE LA RENTABILITE DU CASINO:")
esperance_casino, profit_total = calculer_esperance_casino(1000)

print(f"\n Le casino gagne en moyenne {esperance_casino:.2f} € par partie !")
print(f"Sur 1000 parties, le casino empoche {profit_total:.2f} € !")
```
